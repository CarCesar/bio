```python
from datetime import datetime, timedelta

import ipyplot
from dateutil.relativedelta import relativedelta
from IPython.display import Image
from pyinaturalist import (
    ICONIC_TAXA,
    Observation,
    TaxonCount,
    UserCount,
    enable_logging,
    get_observation_histogram,
    get_observation_identifiers,
    get_observation_observers,
    get_observation_species_counts,
    get_observations,
    pprint,
)
from rich import print

enable_logging()
```

```python
raposas = '''            Cerdocyon thous
            Lycalopex culpaeus
            Lycalopex fulvipes
            Lycalopex griseus
            Lycalopex gymnocercus
            Lycalopex sechurae
            Lycalopex vetulus
            Vulpes lagopus
            Vulpes vulpes
            Vulpes velox
            Vulpes macrotis
            Vulpes corsac
            Vulpes chama
            Vulpes pallida
            Vulpes bengalensis
            Vulpes ferrilata
            Vulpes cana
            Vulpes rueppellii
            Vulpes zerda
            Urocyon cinereoargenteus
            Urocyon littoralis
            Urocyon sp.
            Otocyon megalotis'''.split('\n')
```

```python
raposas[0].strip()
```

```python
%%time
dic = dict()
for a in raposas:
  nome_cientifico = a.strip()
  response = get_observations(taxon_name=nome_cientifico,page='1',photos = True,geo=True,)
  m = Observation.from_json_list(response)
  dic[nome_cientifico] = {'index':raposas.index(a), 'dados': m}
```

```python
dic
```

```python
len(dic)
```

```python
dic1 = dict()
count=1
for k in dic:
  if len(dic[k]['dados']) != 0:
    for a in range(len(dic[k]['dados'])):
      dic1[count] = dict()
      dic1[count]['specie'] = k
      dic1[count]['loc1'] = dic[k]['dados'][a].location[0]
      dic1[count]['loc2'] = dic[k]['dados'][a].location[1]
      dic1[count]['index'] = count#dic[k]['index']
      dic1[count]['image'] = dic[k]['dados'][a].photos[0].url.replace('square','medium')
      dic1[count]['name'] = dic[k]['dados'][a].species_guess
      count+=1
```

```python
dic1
```

```python
import pandas as pd
df1 = pd.DataFrame.from_dict(dic1,orient='index')
```

```python
from vega_datasets import data
import altair as alt

source = alt.topo_feature(data.world_110m.url, 'countries')

world = alt.Chart(source).mark_geoshape(
    fill='white',
    stroke='#aaaaaa',
).project('naturalEarth1')
```
